```python
import matplotlib.pyplot as plt

xmin = -10
ymin = -10
xmax = 10
ymax = 10

fig, ax = plt.subplots()

plt.axis([xmin, xmax, ymin, ymax])

# It is same as draw a line btwn (xmin,0) to (xmax, 0)
plt.plot([xmin, xmax], [0,0], "b")
# Same as drawing a line btwn (0,ymin) to (0,ymax)
plt.plot([0,0], [ymin, ymax], "g")
# plot a point, 4 -> x axis, 5 -> y axis
plt.plot([4],[5], "ro")
# Like this, in "bo", b is blue, o is the marking style.
plt.plot([2],[5], "bo")
# Show the plot
plt.show()
```

# Ploting functions without loop

```python
import matplotlib.pyplot as plt 
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2*(xmax - xmin)
x = np.linspace(xmin, xmax, points) # Points are the total number of points.

fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax]) #Window size
plt.plot([xmin, xmax], [0,0], 'b') #blue x axis
plt.plot([0,0], [ymin, ymax], 'g') #green y axis

y = 2*x + 1
plt.plot(x,y, 'r')

plt.show()
```

# More ways to customize a graph

# LINSPACE IN PYTHON AND MATLAB and INDEXING
1.the linspace function in both python and matlab works same. start, stop, step
      in python it is in numpy library. 
      you will write like 
      import numpy as np
      np.linspace(start, stop, no. of points )
      
      and if you given interval then use np.arange(start, stop, interval)

      in matlab it is direct, 
      a = linspace(start, stop, no. of points)

      here number of points is the total number of points. 
      then it will divide equally btwn that number of points. i.e. linspace -> linearly spaced.

2. but in strng slicing, it is slightly different from MATLAB, 
    [start:stop:interval]
    interval or step

3. but in matlab, matrix can be built by, 
    a = 3:2:10 
    i.e. start:interval:stop
    interval or step


INDEXING 
1. In most languages indexing starts from 0. while in MATLAB  it starts from 1.



```python
import matplotlib.pyplot as plt
import numpy as np

# theses are the maximum and minimum values of the graph
xmin = -10
xmax = 10
ymin = -10
ymax = 10
# Variable for number of points in graph on axis
points = 2*(xmax-xmin)

#This is same as normal sequence
x = np.linspace(xmin, xmax , points) 

# It is same as the subplot in MATLAB, It will also take 3 argument. (row, column, index) of the plot.
# Instead of fig, ax ; We can put any variable. but these are most used conventions. 
fig, ax = plt.subplots()


plt.axis([xmin, xmax, ymin, ymax]) #Window size
plt.plot([xmin,xmax],[0,0],'b') #xaxis
plt.plot([0,0],[ymin, ymax],'g') # yaxis

#this is for giving the labels
ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_title("Graph")
# To show the grid
ax.grid(True)
# you can write True inside bracket or not also.

# Set tick marks on the axes.
# The arrange function works as (start, stop, step)
# step or interval
# Remember
# Every time, inside bracket everywhere, every language, (start, stop, stepORintervalORno. of points)
ax.set_xticks(np.arange(xmin, xmax, 2))
ax.set_yticks(np.arange(ymin, ymax, 2))
        



# equation
y = 2*x + 4 
# plot of the equation
plt.plot(x, y, label='y=2x+1')
#plot for a point (4,6)
plt.plot([-4], [5], 'ro', label='point')
# plot for a line having points (x,3x)
plt.plot(x, 4*x, label='steeper line')

# show the legends for plot
# Default value of the location of legend is loc='best'
# there are also string values for the location like best=0, upper left=1 etc. you can write both location string or string values. 
# better to use string.
# syntax is plt.legend(loc='upper left')
plt.legend()
#show the plot
plt.show()



```
