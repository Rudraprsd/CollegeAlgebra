```python
import matplotlib.pyplot as plt 
import numpy as np 

xmin, xmax = -10, 10
ymin, ymax = -10, 10
points = 100 # Increase number of points for smoother curve
x = np.linspace(xmin, xmax, points)

plt.axis([xmin, xmax, ymin, ymax]) # window size
plt.plot([xmin, xmax],[0,0], 'b') # x- axis
plt.plot([0,0], [ymin, ymax], 'b') # y - axis

# graph for y = c , c = constant
plt.plot([xmin, xmax], [5,5], 'r')

plt.show()
```

## Absolute value graph
### $y = |x|$

```python
import matplotlib.pyplot as plt 
import numpy as np 

xmin, xmax = -10, 10
ymin, ymax = -10, 10
points = 100 # Increase number of points for smoother curve
x = np.linspace(xmin, xmax, points)

plt.axis([xmin, xmax, ymin, ymax]) # window size
plt.plot([xmin, xmax],[0,0], 'b') # x- axis
plt.plot([0,0], [ymin, ymax], 'b') # y - axis

# for absolute value use abs(x), do not have to import any module.
plt.plot(x, abs(x), 'r')

plt.show()
```

## Square root graph
### $y = \sqrt{x}$

```python
import matplotlib.pyplot as plt 
import numpy as np 

xmin, xmax = -10, 10
ymin, ymax = -10, 10
points = 100 # Increase number of points for smoother curve
x = np.linspace(0, xmax, points) # changed for square root as it will only take positive numbers

plt.axis([xmin, xmax, ymin, ymax]) # window size
plt.plot([xmin, xmax],[0,0], 'b') # x- axis
plt.plot([0,0], [ymin, ymax], 'b') # y - axis

#  We will use np.sqrt here, instead of math.sqrt 
plt.plot(x, np.sqrt(x), 'r')
plt.plot(x, -np.sqrt(x), 'g')

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np 
import math
from ipywidgets import interactive

# define the graphing fucniton 
def f(a, zoom):
    xmin = -zoom 
    xmax = zoom 
    ymin = -zoom 
    ymax = zoom 
    points = 100
    x = np.linspace(xmin, xmax, points)
    
    plt.axis([xmin, xmax, ymin, ymax]) # window size
    plt.plot([xmin, xmax], [0,0], 'k') # x-axis
    plt.plot([0,0], [ymin, ymax], 'k') # y-axis
    
    # line 1
    y1 = a
    y2 = a
    plt.plot([xmin, xmax], [y1, y2])
    plt.show()
    
# set up the sliders
interactive_plot = interactive(f, a=(-9, 9), zoom=(1, 100))
interactive_plot
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np 
import math
from ipywidgets import interactive

# define the graphing fucniton 
def f(m, b, zoom):
    xmin = -zoom 
    xmax = zoom 
    ymin = -zoom 
    ymax = zoom 
    points = 100
    x = np.linspace(xmin, xmax, points)
    
    plt.axis([xmin, xmax, ymin, ymax]) # window size
    plt.plot([xmin, xmax], [0,0], 'k') # x-axis
    plt.plot([0,0], [ymin, ymax], 'k') # y-axis
    
    # line 1
    y = m*x + b
    plt.plot(x, y)
    plt.show()
    
# set up the sliders
interactive_plot = interactive(f, m=(-9, 9), b=(-9, 9), zoom=(1, 100))
interactive_plot
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np 
import math
from ipywidgets import interactive

# define the graphing fucniton 
def f(a, b, c, zoom):
    xmin = -zoom 
    xmax = zoom 
    ymin = -zoom 
    ymax = zoom 
    points = 200
    x = np.linspace(xmin, xmax, points)
    
    plt.axis([xmin, xmax, ymin, ymax]) # window size
    plt.plot([xmin, xmax], [0,0], 'k') # x-axis
    plt.plot([0,0], [ymin, ymax], 'k') # y-axis
    
    # line 1
    y = a*x**2 + b*x + c
    plt.plot(x, y)
    plt.show()
    
# set up the sliders
interactive_plot = interactive(f, a=(-9, 9), b=(-9, 9), c=(-9, 9), zoom=(1, 100))
interactive_plot
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np 
import math
from ipywidgets import interactive

# define the graphing fucniton 
def f(a, b, c, zoom):
    xmin = -zoom 
    xmax = zoom 
    ymin = -zoom 
    ymax = zoom 
    points = 200
    x = np.linspace(xmin, xmax, points)
    
    plt.axis([xmin, xmax, ymin, ymax]) # window size
    plt.plot([xmin, xmax], [0,0], 'k') # x-axis
    plt.plot([0,0], [ymin, ymax], 'k') # y-axis
    
    # line 1
    y = a*abs(x-b) + c
    plt.plot(x, y)
    plt.show()
    
# set up the sliders
interactive_plot = interactive(f, a=(-9, 9), b=(-9, 9), c=(-9, 9), zoom=(1, 100))
interactive_plot
```
