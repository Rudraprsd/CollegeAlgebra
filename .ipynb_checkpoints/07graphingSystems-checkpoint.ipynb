## Plotting two equations

```python
import matplotlib.pyplot as plt
import numpy as np 

# Define the range
xmin = -10
xmax = 10
ymin = -10
ymax = 10

# Set number of points to plot
points = 10*(xmax-xmin) 

# Required for plot
plt.axis([xmin, xmax, ymin, ymax]) # Set window size
plt.plot([xmin, xmax],[0,0], 'b') # x-axis
plt.plot([0,0],[ymin, ymax], 'b') # y-axis

# define the array of x value once
x = np.linspace(xmin, xmax, points) 

# Define equations
y1 = 2*x
plt.plot(x, y1, 'r')

y2 = x**2 - 3
plt.plot(x, y2, color='yellow') #remember the syntax for color in matplotlib.

# plot labeling
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('Plot')

# to save the plot
# by default it saves in .png file
plt.savefig("Output1")

# to show the plot
plt.show()
```

## Shading and graphing inequalities

```python
import matplotlib.pyplot as plt
import numpy as np 

xmin = -10
xmax = 10
ymin = -10
ymax = 10

points = 5*(xmax-xmin)

plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax],[0,0], 'b')
plt.plot([0,0],[ymin, ymax], 'b') 

x = np.linspace(xmin, xmax, points)

# line 1
y1 = x + 6
plt.plot(x, y1, color='black')
plt.fill_between(x, y1, 10, facecolor='red') # y1 and 10 are y values.
# line 2
y2 = x + 3
plt.plot(x, y2, color='black')
plt.fill_between(x, y2, y1)
# line 3
y3 = x -1 
plt.plot(x, y3, color='black')
plt.fill_between(x, y3, y2)
# line 4
y4 = x - 4
plt.plot(x, y4, color='black')
plt.fill_between(x, y4, y3)

plt.show()
```

## Interactive Graphs with Sliders

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

def f(m, b, zoom):
    xmin = -zoom
    xmax = zoom 
    ymin = -zoom 
    ymax = zoom
    
    points = 10*xmax - xmin
    x = np.linspace(xmin, xmax, points)
    
    plt.axis([xmin, xmax, ymin, ymax]) # Window Size
    plt.plot([xmin, xmax], [0,0], 'k')
    plt.plot([0,0],[ymin, ymax], 'k')
    plt.title('y = mx + b')
    
    plt.plot(x, m*x + b)
    plt.show()
    
# set a function for slider   
interactive_plot = interactive(f, m=(-9,9), b=(-9,9), zoom=(1,100))
interactive_plot    
```

## Columns:
time, tavg, tmin, tmax, prcp, snow, wdir, wspd, wpgt, pres, tsun

tavg = average temp.
tmin = min. time
tmax = max. time
prcp = precipitation
wdir = wind direction 
wspd = wind speed 
pres = pressure
tsun = amount of sun
wpgt = wind gust ... something


### These are the syntax for the resp. factors.

```python
# Import meteostat library and its dependencies
from datetime import datetime
import matplotlib.pyplot as plt 
from meteostat import Point, Daily

# set time period 
start = datetime(2022, 1, 1) # Year, Month, Date
end = datetime(2022, 7, 31)

# create a point for Berhampur ( for any location )
bpr = Point(19.304420, 84.777069) # latitude , Longitude

# Get daily data
data = Daily(bpr, start, end)
data = data.fetch()

# Plot line chart including average, minimum and maximum temperature.

data.plot(y = ['tavg', 'tmin', 'tmax'])

plt.show()

data.plot(y = ['tavg', 'prcp'])

plt.show()
```
