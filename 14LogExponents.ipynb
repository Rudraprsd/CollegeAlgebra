## Logarithms and Exponents

$$\log_2{2^x} = x $$ 
$$ 2^{log_2{x}} = x $$

### See carefully these relations

### Sometimes we need to find the power, for applications like radioactivity or exponential growth or something.
### eg. 
$$3^x = 25$$
then we take log with same base both side, to find out the x.
$$ \log_3{3^x} = log_3{25}$$
$$x = \frac{\log{25}}{\log{3}}$$

if no base in log then it is $\log_{10}$

ln means the natural log i.e. $\log_e$

## Base 10

```python
import math
print(math.log(10000, 10))
# python weirdness 
print(math.log(1000, 10))
# so write like this instead, 
print(round(math.log(1000, 10)))
```

## Natural log and other bases

```python
import math
print(math.log(16, 2))
print(math.log(math.e**5))
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
import math

xmin, xmax = -10, 10
ymin, ymax = -10, 10
points = 100 # Increase number of points for smoother curve
x = np.linspace(xmin, xmax, points)
x2 = np.linspace(.01, xmax, points)

plt.axis([xmin, xmax, ymin, ymax]) # window size
plt.plot([xmin, xmax],[0,0], 'b') # x- axis
plt.plot([0,0], [ymin, ymax], 'b') # y - axis

# graph for y = c , c = constant
y1 = math.e**x
y2 = x
y3 = np.log(x2)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x2, y3)

# Don't do like this, 
# plt.plot([x, x, x2],[y1, y2, y3])
# It has a lot different meaning

plt.show()
```
