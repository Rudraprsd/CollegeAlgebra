## Solving equation with two Variable

```python
# this means from sympy, import everything.
from sympy import * 

# Define the variable by giving symbols 
x, y = symbols('x y')

# Equations set equal to zero. 
first = 2*x + y - 1
second = x - 2*y + 7

# The solution 
print(linsolve([first, second], (x, y)))

# OR
print(linsolve([first, second], [x, y]))


```

```python
# this means from sympy, import everything.
from sympy import * 

# Define the variable by giving symbols 
x, y = symbols('x y')

# Equations set equal to zero. 
first = 2*x + y - 1
second = x - 2*y + 7

solution = linsolve([first, second], [x, y])
# important - ----------------------------
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]

print(f"({x_solution}, {y_solution})")
```

# Solving Linear Equations in two variables
## Factor to change from $'0='$ to $'y='$ preparing to graph

```python
from sympy import * 
from sympy.plotting import plot
# By using the plotting library of sympy

var('x y')

# Equations set equal to zero. 
first = 2*x + y - 1
second = x - 2*y + 7

solution = linsolve([first, second], [x, y])

# important - ----------------------------
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]

print(f"({x_solution}, {y_solution})")

# Sympy syntax for equation equals to 0, ready to factor.
# y_first = Eq(first, 0 ) 
# We can use the above commented statement, but our equation already set to zero. 

#sympy solve for y 
y1 = solve(first, y)

# same steps for second equation 
# y_second = Eq(second, 0)
y2 = solve(second, y)

# show factored results
print(f"y1 = {y1[0]}")
print(f"y2 = {y2[0]}")

# Plotting
# x = symbols('x')
# Don't need
plot(y1[0], y2[0], (x, -10, 10))
```
