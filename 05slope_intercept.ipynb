# Intercept

```python
def intercept(x1, x2, y1, y2):
    slope = (y2 - y1)/(x2 - x1)
    c = y1 - slope*x1
    print(f"Full equation is y = {slope}*x + {c}")
    print(f"Intercept is {c}")

x1 = int(input("Enter x1 = "))
x2 = int(input("Enter x2 = "))
y1 = int(input("Enter y1 = "))
y2 = int(input("Enter y2 = "))

intercept(x1, x2, y1, y2)
```
